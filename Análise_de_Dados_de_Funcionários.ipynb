```python
# Importando bibliotecas para análise e visualização de dados
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregando o arquivo CSV em um DataFrame
# Certifique-se de substituir pelo caminho correto do arquivo
arquivo = '/content/funcionarios.csv'  # Altere conforme necessário
arq = pd.read_csv(arquivo)

# Exibir as primeiras linhas para visualizar a estrutura do dataset
display(arq.head())

# Verificar informações gerais do DataFrame
arq.info()

# Estatísticas descritivas das colunas numéricas
display(arq.describe())

# Verificar valores nulos e possíveis problemas nos dados
print("Valores nulos por coluna:")
print(arq.isnull().sum())

# Converter a coluna de data de admissão para o tipo datetime (se existir no dataset)
if 'Data_Admissao' in arq.columns:
    arq['Data_Admissao'] = pd.to_datetime(arq['Data_Admissao'], errors='coerce')

# Criando um gráfico de distribuição salarial (se a coluna existir)
if 'Salario' in arq.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(arq['Salario'], bins=20, kde=True)
    plt.title('Distribuição Salarial')
    plt.xlabel('Salário')
    plt.ylabel('Frequência')
    plt.show()

# Criando um gráfico para analisar a quantidade de admissões ao longo do tempo
if 'Data_Admissao' in arq.columns:
    plt.figure(figsize=(12, 6))
    arq['Ano_Admissao'] = arq['Data_Admissao'].dt.year
    sns.countplot(data=arq, x='Ano_Admissao', order=sorted(arq['Ano_Admissao'].dropna().unique()))
    plt.title('Quantidade de Admissões por Ano')
    plt.xlabel('Ano de Admissão')
    plt.ylabel('Quantidade de Funcionários')
    plt.show()

# Insight: Podemos analisar se houve aumento ou redução nas contratações ao longo do tempo
```
